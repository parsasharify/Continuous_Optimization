```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
from tqdm import tqdm
import pandas as pd
from collections import namedtuple
import random
```

```python
import math
def f1(x):
    return np.exp(abs(x**10 - x**8))
    pass

def f2(x):
    if(x < 0):
      return None
    
    return np.exp(x**2 + 2*x + 1/x)
    pass

def f3(x):
    return -1 * np.log(abs(2 * np.cos(x) + 100/x))
    pass
```

```python
ConvexityResult = namedtuple("ConvexityResult", ["is_convex", "exc_interval", "exc_mid_pt"])

def check_convexity(func):
    # Check the input function for convexity
    is_convex = True
    exc_start_pt = None
    exc_end_pt = None
    exc_mid_pt = None
    for start in np.linspace(0, 10, num = 100):
      end = start + 0.1
      for t in np.linspace(0, 1, num = 10):
        if(func(t * start + (1-t) * end) > t * func(start) + (1-t) * func(end)):
          is_convex = False
          exc_start_pt = start
          exc_end_pt = end
          exc_mid_pt = t * start + (1-t) * end
          break
      if(not is_convex):
        break
    res = ConvexityResult(is_convex, (exc_start_pt, exc_end_pt), exc_mid_pt)
    # if is_convex == False then exc values are None
    return res
```

```python
def visualize(func, convexity_res):
    if(convexity_res[0]):
      plt.plot(np.linspace(0, 10, num = 100), [func(x) for x in np.linspace(0, 10, num = 100)], color ='tab:blue')
    else:
      
      plt.plot(np.linspace(0, convexity_res[1][0], num = 100), [func(x) for x in np.linspace(0, convexity_res[1][0], num = 100)], color ='tab:blue')
      plt.plot(np.linspace(convexity_res[1][0], convexity_res[1][1], num = 100), [func(x) for x in np.linspace(convexity_res[1][0], convexity_res[1][1], num = 100)], color ='tab:red')
      plt.plot(np.linspace(convexity_res[1][1], 10, num = 100), [func(x) for x in np.linspace(convexity_res[1][1], 10, num = 100)], color ='tab:blue')

    plt.show()


    pass
```

```python
visualize(f1, check_convexity(f1))
visualize(f2, check_convexity(f2))
visualize(f3, check_convexity(f3))
```

```python

def g(x, y):
    return np.cos(x)*np.cos(y)*np.exp(-x**2-y**2)

def f(x, y):
    return -g(2*x,2*y) + \
        0.05*g(2*x + 0.5, 2*y + 0.5) - 0.1*g(x + 1, y + 1) + \
        0.05*g(2*x + 0.5, 2*y - 0.5) - 0.1*g(x + 1, y - 1) + \
        0.05*g(2*x - 0.5, 2*y + 0.5) - 0.1*g(x - 1, y + 1) + \
        0.05*g(2*x - 0.5, 2*y - 0.5) - 0.1*g(x - 1, y - 1)


```

```python

def grad(func, x, y, eps=1e-8):
    x_grad = (func(x + eps , y) - func(x - eps , y))/(2*eps)
    y_grad = (func(x , y + eps) - func(x , y - eps))/(2*eps)
    return (x_grad, y_grad)
```

```python
x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
def path_visualize(pts):

    global X, Y, Z

    
    # pts: [N, 2]
    # visualizes input array of points constructing convergence path
    plt.figure(figsize=(10, 10))
    plt.contour(X, Y, np.exp(Z)**10, cmap='Spectral')
    a_x, a_y = np.array(pts).T
    plt.plot(a_x, a_y, color='red');


```

```python

def GD(func, init_x, init_y, lr, max_iter):
    # Perform gradient descent update iterations
    rows, cols = (max_iter+1, 2)
    x_list = []
    y_list = []
    x_list.append(init_x)
    y_list.append(init_y)
    record =[[0]*cols]*rows
    record[0][0] = init_x
    record[0][1] = init_y
    for i in range(max_iter):
        #print(grad(func , init_x , init_y , 1e-8))
        diff_x = -1 * lr * grad(func , init_x , init_y , 1e-8)[0]
        diff_y = -1 * lr * grad(func , init_x , init_y , 1e-8)[1]
        init_x += diff_x
        init_y += diff_y
        print(init_x)
        print(init_y)
        x_list.append(init_x)
        y_list.append(init_y)
        record[i+1][0] = init_x.copy()
        record[i+1][1] = init_y.copy()
    print(y_list)
    return np.transpose(np.array([x_list , y_list]))
    pass

def minimize(func, lr, max_iter=10000):
    # minimize the given function and visaulize the results
    for i in range(len(lr)):
        for j in range(5):
            value_x = random.randint(-3, 3)
            value_y = random.randint(-3, 3)
            #value_x = 1.5
            #value_y = 1.5
            record = GD(func, value_x, value_y, lr[i], max_iter)
            path_visualize(record)
            print(record)
            print(f"initial x = {value_x}\n"
                  f"initial y = {value_y}\n"
                  f"lr = {lr[i]}\n")
    pass



```

```python
def GD_variable_lr(func, init_x, init_y, lr_init, max_iter):
    decay_rate = lr_init / max_iter

    rows, cols = (max_iter + 1, 2)
    record = [[0] * cols] * rows
    x_list = []
    y_list = []
    x_list.append(init_x)
    y_list.append(init_y)
    record[0][0] = init_x
    record[0][1] = init_y
    for i in range(max_iter):
        diff = -1 * lr_init * grad(func, init_x, init_y, 1e-8)

        init_x += diff
        init_y += diff
        x_list.append(init_x)
        y_list.append(init_y)
        record[i+1][0] = init_x
        record[i+1][1] = init_y
        lr_init -= decay_rate

    return np.transpose(np.array([x_list , y_list]))

    pass

def minimze_variable_lr(func, lr_init, max_iter=10000):
    # minimze the given function and visaulize the results
    for j in range(5):
        value_x = random.randint(-3, 3)
        value_y = random.randint(-3, 3)
        record = GD(func, value_x, value_y, lr_init, max_iter)
        path_visualize(record)
        print(f"initial x = {value_x}\n"
              f"initial y = {value_y}\n")

    pass

```

```python

def GD_momentum(func, init_x, init_y, lr, max_iter):
    # add momentum to GD update equation
    delta_w_x = 0
    delta_w_y = 0
    alpha = .9
    rows, cols = (max_iter + 1, 2)
    x_list = []
    y_list = []
    x_list.append(init_x)
    y_list.append(init_y)
    record = [[0] * cols] * rows
    record[0][0] = init_x
    record[0][1] = init_y
    for i in range(max_iter):

        beta = .9
        delta_w_x = -1 * lr * (beta * delta_w_x + (1 - beta) * grad(func, init_x, init_y, 1e-8)[0])       #alpha * delta_w_x - lr * grad(func, init_x, init_y, 1e-8)[0]
        delta_w_y = -1 * lr * (beta * delta_w_y + (1 - beta) * grad(func, init_x, init_y, 1e-8)[1])

        init_x += delta_w_x
        init_y += delta_w_y
        x_list.append(init_x)
        y_list.append(init_y)
        record[i+1][0] = init_x
        record[i+1][1] = init_y


    return np.transpose(np.array([x_list , y_list]))
    pass

def minimze_momentum(func, lr, max_iter=10000):
    # minimze the given function and visaulize the results
    for j in range(5):
        value_x = random.randint(-3, 3)
        value_y = random.randint(-3, 3)
        value_x = 1
        value_y = 2
        record = GD_momentum(func, value_x, value_y, lr, max_iter)
        path_visualize(record)
        print(f"initial x = {value_x}\n"
              f"initial y = {value_y}\n")

    pass

```
